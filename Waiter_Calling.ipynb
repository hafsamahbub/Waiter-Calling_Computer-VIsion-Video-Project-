```python
# Image Stitching
# Here we combine the selected frames into a 4x2 grid for visualization.
from PIL import Image
import matplotlib.pyplot as plt

# List of uploaded image filenames
image_files = ['no.1.png', 'no.2.png', 'no.3.png', 'no.4.png',
               'no.5.png', 'no.6.png', 'no.7.png', 'no.8.png']

# Open all images
images = [Image.open(img) for img in image_files]

# Find smallest width and height to make all images uniform
min_width = min(img.width for img in images)
min_height = min(img.height for img in images)

# Resize all images to the same size
resized_images = [img.resize((min_width, min_height)) for img in images]

# Define grid size (4 columns, 2 rows)
rows = 2
cols = 4

# Create a blank canvas for the grid
grid_image = Image.new('RGB', (cols * min_width, rows * min_height))

# Paste each image into the correct position in the grid
for index, img in enumerate(resized_images):
    x = (index % cols) * min_width
    y = (index // cols) * min_height
    grid_image.paste(img, (x, y))

# ✅ Display the stitched image correctly inside Colab
plt.figure(figsize=(12, 6))
plt.imshow(grid_image)
plt.axis('off')
plt.show()

# Save the final stitched image
grid_image.save('stitched_grid.png')

```
